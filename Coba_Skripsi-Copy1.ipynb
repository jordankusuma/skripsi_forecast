```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df_sales = pd.read_csv('proyek_skripsi/sales.csv')
df_sales
```

```python
df_sales.info()
```

```python
df_sales.nunique()
```

```python
#drop column that has > 80% null value
df_sales = df_sales.drop(['note', 'staff_note', 'order_discount_id', 'updated_by', 'updated_at', 'return_id', 'attachment', 'return_sale_ref', 'sale_id', 'rounding', 'suspend_note', 'address_id', 'reserve_id', 'salesman_id', 'salesman_commission', 'total_commission', 'product_discount', 'total_discount', 'grand_total', 'payment_term', 'due_date', 'paid', 'customer', 'quote_id', 'reference_no', 'biller_id', 'biller', 'warehouse_id', 'order_discount', 'product_tax', 'order_tax_id', 'order_tax', 'total_tax', 'shipping', 'pos', 'surcharge', 'api', 'shop', 'hash', 'created_by', 'sale_status', 'payment_status', 'return_sale_total', 'total_items'], axis=1)
```

```python
#df_sales.describe()
```

```python
df_sales = df_sales[(df_sales[['total']] >= 0).all(1)]
```

```python
df_sales = df_sales.rename(columns={'id': 'sale_id'})
```

```python
df_sales.info()
```

```python
df_sales = df_sales.loc[(df_sales['date'] >= '2017-02-31')]
```

```python
# df_sales.loc[(df_sales['date'] <= '2017-02-31')]
```

```python
df_detail = pd.read_csv('proyek_skripsi/sale_items.csv')
df_detail
```

```python
df_detail.info()
```

```python
df_detail.nunique()
```

```python
#remove unnessary column
df_detail.drop(['serial_no', 'sale_item_id', 'comment', 'id', 'product_type', 'option_id', 'unit_price', 'net_unit_price', 'warehouse_id', 'item_tax', 'tax_rate_id', 'tax', 'subtotal', 'product_unit_id', 'product_unit_code', 'unit_quantity'], inplace=True, axis=1)
```

```python
df_detail.describe()
```

```python
df_detail = df_detail[(df_detail[['quantity', 'item_discount', 'real_unit_price']] >= 0).all(1)]
```

```python
df_detail = df_detail.fillna(0)
```

```python
df_detail.info()
```

```python
df_result = pd.merge(df_sales, df_detail, on='sale_id', how='outer')
```

```python
df_result.info()
```

```python
df_result = df_result.dropna()
```

```python
df_result['date'] = pd.to_datetime(df_result['date'], errors='coerce')
```

```python
cols = ['total', 'quantity', 'real_unit_price']
df_result[cols] = df_result[cols].applymap(np.int64)
```

```python
df_result.info()
```

```python
df_forecast_month = df_result[["date", "quantity", "real_unit_price"]]
```

```python
#df_forecast_date = df_forecast_month
```

```python
df_forecast_month
```

```python
df_forecast_month.loc[(df_forecast_month['date'] <= '2017-3-31')]
```

```python
period = df_forecast_month.date.dt.to_period("M")
```

```python
df_forecast_month = df_forecast_month.groupby(period).mean()
```

```python
#inflasi dan IHK
# df_2017 = df_forecast_date[(df_forecast_date['date'] >= '2017-01-01') & (df_forecast_date['date'] < '2018-01-01')]
df_ihk_inflasi = pd.read_csv("proyek_skripsi/Data_IHK_Inflasi_Nasional.csv")
```

```python
df_ihk_inflasi['date'] = pd.to_datetime(df_ihk_inflasi['date'], errors='coerce')
```

```python
df_ihk_inflasi.head()
```

```python
df_forecast_month = pd.merge(df_forecast_month, df_ihk_inflasi, on='date', how='outer')
```

```python
df_forecast_month.set_index('date')
```

```python
plt.figure(figsize=(20,5))
plt.plot(df_forecast_month['date'], df_forecast_month['IHK'], label="IHK")
plt.title("Chart of IHK")
plt.xlabel=("date")
plt.ylabel =("IHK")
plt.legend(loc = "upper left")
plt.show()
#df_forecast_month.set_index('date').plot(figsize=(10,5), grid=True)
```

```python
plt.figure(figsize=(20,5))
plt.plot(df_forecast_month['date'], df_forecast_month['Inflasi'], label="Inflasi")
plt.title("Chart of Inflasi")
plt.xlabel=("date")
plt.ylabel =("Inflasi")
plt.legend(loc = "upper left")
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(df_forecast_month['date'], df_forecast_month['quantity'], label='quantity')
plt.title("Chart of quantity")
plt.xlabel=("Date")
plt.ylabel =("Quantity")
plt.legend(loc = "upper left")
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(df_forecast_month['date'], df_forecast_month['real_unit_price'], label='price')
plt.title("Chart of price")
plt.xlabel=("Date")
plt.ylabel =("Price")
plt.legend(loc = "upper left")
plt.show()
```
